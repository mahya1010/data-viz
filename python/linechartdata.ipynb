```python
import pandas as pd
```

```python

file_path = '../data/Cleaned_Life_Expectancy_Data.csv'
```

```python
# Load the dataset
data = pd.read_csv(file_path)
data.head()
```

```python
# Grouping by year and calculating the average life expectancy for each year
average_life_expectancy_per_year = data.groupby('Year')['Life expectancy '].mean().reset_index()
```

```python
average_life_expectancy_per_year.head()
```

```python
# Sorting the data for better visualization
average_life_expectancy_per_year = average_life_expectancy_per_year.sort_values(by= 'Year')
```

```python
average_life_expectancy_per_year.head()
```

```python
#Save this to a new CSV file
average_life_expectancy_per_year.to_csv('../data/average_life_expectancy_per_year.csv', index=False)
```
