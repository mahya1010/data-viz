```python
import pandas as pd
```

```python

file_path = '../data/Cleaned_Life_Expectancy_Data.csv'
```

```python
# Load the dataset
data_full = pd.read_csv(file_path)
data_full.head()
```

```python
# Define life expectancy ranges
bins = [0, 20, 30, 40, 50, 70, 100]
labels = ['0-20', '21-30', '31-40', '41-50', '51-70', '71-100']
```

```python
# Categorize life expectancy into ranges
data_full['Life Expectancy Range'] = pd.cut(data_full['Life expectancy '], bins=bins, labels=labels, right=False)
```

```python
data_full.head()
```

```python
# Group by year and life expectancy range, and count the number of entries in each group
grouped_data = data_full.groupby(['Year', 'Life Expectancy Range']).size().reset_index(name='Count')
```

```python
grouped_data.head()
```

```python
#Save this to a new CSV file
grouped_data.to_csv('../data/life_expectancy_per_year_waffle.csv', index=False)
```
